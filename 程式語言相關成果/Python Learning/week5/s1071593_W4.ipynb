```python
# encoding=utf-8
# programmer: 黃駿瑜
# Date: 2020/3/31
# Python 整數型態: int vs. float

myInt = 10    # 沒有小數點的數值為整數(int)
myFloat = 10    # 有小數點的數值為浮點數(float)

print("整數",type(myInt),myInt)
print("浮點數",type(myFloat),myFloat)
```

```python
# encoding=utf-8
# programmer: 黃駿瑜
# Date: 2020/3/31
# Python 整數型態: int大小與範圍

import sys

myInt8 = 255    # 一般整數大小為28 bytes
myInt16 = 65535    # 但python 3 的整數大小無限制
myGoogol = 10 ** 100    # 1 古戈爾(googol) = 1e100

print("整數",type(myInt8),myInt8,sys.getsizeof(myInt8))
print("整數",type(myInt16),myInt16,sys.getsizeof(myInt16))
print("整數",type(myGoogol),myGoogol,sys.getsizeof(myGoogol))
```

```python
# encoding=utf-8
# programmer: 黃駿瑜
# Date: 2020/3/31
# Python 整數型態: int大小與範圍

import sys    # sys.getsizeof() 可以取得資料的大小, 單位為byte

myInt = 1
for i in range(15):
    myInt *=16 
    print("整數256^",i+1,"次方 = ",myInt," 大小 :",sys.getsizeof(myInt))
```

```python
# encoding=utf-8
# programmer: 黃駿瑜
# Date: 2020/3/31
# Python 整數型態: 2、8、16 進制整數

myBin = 0b11110000    # 2進制整數
myOct = 0o77776543    # 8進制整數
myHex = 0xffff12ad    # 16進制整數

print("2進制整數",type(myBin),myBin)
print("8進制整數",type(myOct),myOct)
print("16進制整數",type(myHex),myHex)
```

```python
# encoding=utf-8
# programmer: 黃駿瑜
# Date: 2020/3/31
# Python 整數型態: 2、8、16 進制整數

myInt1 = 64    # 2進制整數
myInt2 = 1024    # 8進制整數
myInt3 = 10 ** 100    # 16進制整數

# myInt1 2進制、8進制、16進制轉換
print("64的2進制、8進制、16進制轉換: ")
print(bin(myInt1))
print(oct(myInt1))
print(hex(myInt1))

# myInt2 2進制、8進制、16進制轉換
print("1024的2進制、8進制、16進制轉換: ")
print(bin(myInt2))
print(oct(myInt2))
print(hex(myInt2))

# myInt3 2進制、8進制、16進制轉換
print("1 Googol的2進制、8進制、16進制轉換: ")
print(bin(myInt3))
print(oct(myInt3))
print(hex(myInt3))
```

```python
# encoding=utf-8
# programmer: 黃駿瑜
# Date: 2020/3/31
# Python 浮點數型態: float大小與範圍

import sys    # sys.getsizeof() 可以取得資料的大小, 單位為byte

myFloat = 1.0
maxFloat = sys.float_info.max    # sys.float_info.max 取得浮點數最大值
minFloat = sys.float_info.min    # sys.float_info.min 取得浮點數最小值
print("浮點數1 = ",myFloat,type(myFloat))
print("浮點數最大值 = ",maxFloat,type(maxFloat))
print("浮點數最小值 = ",minFloat,type(minFloat))
print("="*60)    # 列印分隔線

for i in range(1024):
    myFloat *= 2 
    print("浮點數 2^",i+1,"次方 = ",myFloat," 大小 :",sys.getsizeof(myFloat))
```
